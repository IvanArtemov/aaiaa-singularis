```python
import fitz  # PyMuPDF for PDF parsing
import spacy

import re
```

```python
!python -m spacy download en_core_web_sm
```

```python
def parse_pdf(path):
    doc = fitz.open(path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text
```

```python
nlp = spacy.load("en_core_web_sm")
paper = "paper.pdf"
```

```python
raw_text = parse_pdf(paper)
doc = nlp(raw_text)
```

```python
# Regex for hypothesis patterns
hypothesis_patterns = [
    r"(hypothesis|H\d+):",
    r"(theor(ies|y)|hypothes[ei]s)\s+(is|was|are|were|that)",
    r"(we|I|authors?|it is|study|theor(ies|y)|hypothes[ei]s)\s+(hypothesi[sz]e|predict|propose|assume|expect|anticipate|foresee|postulate|conjecture|test|demonstrate|show|suggest)",
    r"(hypothesi[sz]e|predict|propose|assume|expect|anticipate|foresee|postulate|aim|conjecture|test|demonstrate|show|suggest)(s|d)?\s+that"
    r"(investigate|examine|predict|test|demonstrate|show)(s|d)?\s+(whether|if)",
    r"(evidence|data)\s+(predict|suggest|propose|assume)"
    r"(aim|objective|goal)\s+(of\s+th(is|e)\s+study)",
    r"if.*then",
    r"it would appear",
    r"it likely that",
    r"According to the (theor(ies|y)|hypothes[ei]s|views?)"
]
compiled_patterns = [re.compile(p, re.IGNORECASE) for p in hypothesis_patterns]
```

```python
hypotheses = []
for ind, sent in enumerate(doc.sents):
    sent_text = sent.text.strip().replace("\n", " ")
    if sent_text and any(pattern.search(sent_text) for pattern in compiled_patterns):
            hypotheses.append(sent_text)

hypotheses
```
